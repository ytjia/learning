```python
poi_consume['consumescore'] = poi_consume.consumecnt / pow(poi_consume.days + 2, 1.8)poi_consume.head(5)
```

```python
poi_consume['consumescore'] = poi_consume.consumecnt / pow(poi_consume.days + 2, 1.8)poi_consume.head(5)
```

```python
# 多列选取poi_cons_info = pd.DataFrame(poi_consume[poi_consume.cityid == 1], columns=['mainpoiid', 'consumescore'])poi_cons_info.describe()
```

```python
grouped = poi_cons_info.groupby(['mainpoiid'])poi_summary = grouped.aggregate(np.sum)poi_summary.head(5)
```

```python
poi_summary.ix[:, 'consumescore'].plot(kind='kde')
```

```python
poi_summary['finalscore'] = log10(poi_summary.consumescore + 1)# poi_summary['finalscore'] = (poi_summary.consumescore - mean(poi_summary.consumescore)) / std(poi_summary.consumescore)# poi_summary['log10finalscoreprice'] = log10(poi_summary.finalscore)poi_summary.describe()
```

```python
poi_summary['01score'] = (poi_summary.finalscore - min(poi_summary.finalscore)) /  (max(poi_summary.finalscore) -  min(poi_summary.finalscore))poi_summary.head(5)
```
